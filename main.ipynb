1. Interpretability
Easy to Understand: MAPE expresses the error as a percentage, making it easy to interpret. For example, a MAPE of 5% means the forecasted price is, on average, 5% away from the actual price.
Relative Error: Since MAPE is a relative metric, it normalizes the error based on the actual values, which makes it easier to compare across different time periods, stocks, or assets with varying price scales.
2. Focus on Prediction Accuracy
MAPE gives you an indication of how close the forecast is to the actual value on average. Since stock price forecasting involves predicting future values, MAPE highlights the magnitude of the prediction error in percentage terms.
3. Widely Used in Business and Economics
MAPE is a common metric in business forecasting (including stock prices), so it is easily understood by stakeholders and can be useful for comparing the performance of different forecasting models.
4. Comparability Across Datasets
Since it is percentage-based, MAPE allows you to compare models across stocks with different price ranges. A model that gives a lower MAPE on one stock can be considered better in terms of accuracy compared to one with a higher MAPE, regardless of the actual price of the stock.
5. Account for Relative Size of the Error
Unlike metrics like Mean Squared Error (MSE), which penalize large errors more significantly due to squaring, MAPE provides a more balanced measure of error, making it more suitable when you need to treat all errors equally in terms of their percentage difference from the actual value.
